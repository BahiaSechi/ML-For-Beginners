# Linear Regression - Diabetes dataset

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
import pandas as pd

X, y = datasets.load_diabetes(return_X_y=True)

print(X.shape)
print(X[0])

df = pd.DataFrame(X, columns=["age", "sex", "bmi", "bp", "tc", "ldl", "hdl", "tch", "ltg", "glu"])

print(df)

# BMI
# X = X[:, np.newaxis, 2]
# Age
# X = X[:, np.newaxis, 0]
# Glu
X = X[:, np.newaxis, 9]

print(X.shape)
print(X[0])

# Splitting training and testing data 
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)

model = linear_model.LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# Displaying the plot
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
# plt.xlabel('Scaled BMIs')
# plt.xlabel('Scaled Ages')
plt.xlabel('Scaled Glu')
plt.ylabel('Disease Progression')
# plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
# plt.title('A Graph Plot Showing Diabetes Progression Against Age')
plt.title('A Graph Plot Showing Diabetes Progression Against Blood Sugar Level')
plt.show()
```
